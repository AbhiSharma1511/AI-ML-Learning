```python
 ## parent class

class Car():
    def __init__(self, windows, doors, engineType):
        self.windows = windows
        self.doors = doors
        self.engineType = engineType

    def drive(self):
        print(f"The person will drive the car with {self.engineType} engine.")
```

```python
car1 = Car(4, 5, "Petrol")
car1.drive()

car2 = Car(4, 7, "Diesel")
car2.drive()
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, enginType, isSelfDriving):
        super().__init__(windows, doors, enginType)
        self.isSelfDriving = isSelfDriving

    def autoDrive(self):
        print(f"Tesla supports self driving: {self.isSelfDriving}")
```

```python
tesla1 = Tesla(4, 5, "Electric", True)
tesla1.drive()
tesla1.autoDrive()
```
