```python
# import the rewuired libraries
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# read the data from the csv file
dataset  = pd.read_csv("data.csv")
```

```python
# print the head, tail and columns of the dataset
print(dataset.head(5))
print(dataset.tail(5))
print(dataset.columns)
```

```python
# divide the dataset into dependent (Y) and independent(X) category
X = dataset.drop("Survived", axis=1)
Y = dataset['Survived']
print(X)
print(Y)
```

```python
# here we will setup the data transformation code
# transformation on x daatset
data_encoder  = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ['Sex','Embarked'])], remainder='passthrough')
X = data_encoder.fit_transform(X)
```
